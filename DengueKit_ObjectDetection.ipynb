```python
import os
path = '/content/darknet/data/obj/train'
images_list_tr=[]
for files in os.listdir(path):
    if files.endswith(('.jpg','.jpeg','.png','JPG')):
        f_n = files
        images_list_tr.append('data/obj/train/{}'.format(f_n))

path2 = '/content/darknet/data/obj/test'
images_list_te=[]
for files in os.listdir(path2):
    if files.endswith(('.jpg','.jpeg','.png','JPG')):
        f_n = files
        images_list_te.append('data/obj/test/{}'.format(f_n))
```

```python
with open("data/train.txt", "w") as f:
    f.write("\n".join(images_list_tr))

with open("data/test.txt", "w") as f:
    f.write("\n".join(images_list_te))
```
